# read csv

```python
import pandas as pd
import os
from glob import glob
```

```python
df_morpheme_train = pd.read_csv('info_morpheme_train.csv')
df_morpheme_val = pd.read_csv('info_morpheme_val.csv')
```

# Simple EDA

```python
# total kind of morpheme label
train_label = df_morpheme_train['label'].unique()
print("There are {} kind of train morpheme labels.".format(len(train_label)))
val_label = df_morpheme_val['label'].unique()
print("There are {} kind of val morpheme labels.".format(len(val_label)))


dupli_label = list(set(train_label).intersection(val_label))
print("There are total of {} labels that exist in both train and validation datasets.".format(len(dupli_label)))
```

```python
print("We need to select target morpheme labels in the {} labels that exist in both train and validation datasets".format(len(dupli_label)))
```

# Find the top label with the largest amount of train data

```python
label_largest_order = list(df_morpheme_train['label'].value_counts().index)
label_largest_order
```

```python
# drop the label if it is not exist in both train and validation datasets
dupli_label_largest_order = []
for label in label_largest_order:
    if label in dupli_label:
        dupli_label_largest_order.append(label)
        
print(len(label_largest_order), len(dupli_label_largest_order))
```

```python
dupli_label_largest_order
```

# define number of top label & find videos

```python
num_top = 100
target_label = dupli_label_largest_order[:num_top]
print(target_label)
```

```python
# train Front video 
# target_dic = {}
target_video = []
for label in target_label:
#     target_dic[label] = df_morpheme_train[df_morpheme_train['label'] == label].copy()
    df = df_morpheme_train[df_morpheme_train['label'] == label].copy()
    target_video += list(df['video_name'])

print(len(target_video), len(set(target_video)))
# target_dic
```

```python
# get only front video
front_video = []
for video in set(target_video):
    if video.split('_')[4].split('.')[0] == 'F':
        front_video.append(video)

print(len(front_video), len(set(front_video)))
```

```python
import os
from glob import glob
import shutil

def copy_file(type_, file_name):
    dir_data = os.path.join('/data', 'KSL', type_)
    if type_ == "Training":
        file_list = glob(os.path.join(dir_data, '*', file_name))
    elif type_ == "Validation":
        file_list = glob(os.path.join(dir_data, '*', '*',  file_name))
    
    if len(file_list) == 0:
        print("try again", file_name)
        file_list = glob(os.path.join(dir_data, '*', '*',  file_name))
        if len(file_list) == 0:
            print("try again", file_name)
            file_list = glob(os.path.join(dir_data, "**", file_name), recursive=True)
    
    if len(file_list) > 1:
        print("more than 1", file_name)
    elif len(file_list) == 0:
        print("There is no file.", file_name)
    else:
        to_path = os.path.join('/data', 'sub_KSL', type_, 'video')
        if not os.path.isdir(to_path):
            os.makedirs(to_path)
        shutil.copy(file_list[0], to_path)
        if not os.path.isfile(file_list[0]):
            print("???", file_name)
```

# check moved video

```python
sub_data_dir = os.path.join('/data', 'sub_KSL')
sub_train_dir = os.path.join(sub_data_dir, 'Training')
sub_val_dir = os.path.join(sub_data_dir, 'Validation')
```

```python
train_video_list = glob(os.path.join(sub_train_dir, 'video', '*.mp4'))
val_video_list = glob(os.path.join(sub_val_dir, 'video', '*.mp4'))
len(train_video_list), len(val_video_list)
```

```python
train_video_name = [os.path.basename(path) for path in train_video_list]
val_video_name = [os.path.basename(path) for path in val_video_list]
```

```python
from tqdm import tqdm

for i, label in tqdm(enumerate(target_label)):
    if i == 0:
        df = df_morpheme_train[df_morpheme_train['label'] == label].copy()
    else:
        df = pd.concat([df, df_morpheme_train[df_morpheme_train['label'] == label].copy()], axis=0) 
        
df_sub_train = df.copy()
```

```python
df_sub_train.reset_index(inplace=True, drop=True)
train_del_i = []
for i in tqdm(range(len(df_sub_train))):
    if not df_sub_train.iloc[i]['video_name'] in train_video_name :
        train_del_i.append(i)
```

```python

print(len(df_sub_train))
df_sub_train.drop(train_del_i, axis=0, inplace=True)
print(len(df_sub_train))
```

```python
print(len(df_sub_train['label'].unique()))
print(len(df_sub_train['video_name'].unique()))
df_sub_train.to_csv(os.path.join('/data', 'sub_KSL', 'Training', 'info.csv'), index=False)
```

```python
for i, label in tqdm(enumerate(target_label)):
    if i == 0:
        df = df_morpheme_val[df_morpheme_val['label'] == label].copy()
    else:
        df = pd.concat([df, df_morpheme_val[df_morpheme_val['label'] == label].copy()], axis=0) 
        
df_sub_val = df.copy()
```

```python
df_sub_val.reset_index(inplace=True, drop=True)
val_del_i = []
for i in tqdm(range(len(df_sub_val))):
    if not df_sub_val.iloc[i]['video_name'] in val_video_name :
        val_del_i.append(i)
```

```python
print(len(df_sub_val))
df_sub_val.drop(val_del_i, axis=0, inplace=True)
print(len(df_sub_val))
```

```python
print(len(df_sub_val['label'].unique()))
print(len(df_sub_val['video_name'].unique()))
df_sub_val.to_csv(os.path.join('/data', 'sub_KSL', 'Validation', 'info.csv'), index=False)
```

# Fine the demo test video

```python
# define unique validaiton video
list_val_video = list(df_sub_val['video_name'].unique())
len(list_val_video)
```

```python
target_demo_video = []

for video in list_val_video:
    label_in_video = list(df_sub_val[df_sub_val['video_name'] == video]['label'])
    
    
    if len(label_in_video) > 6:
        
        cnt = 0
        for label in label_in_video:
            if label in target_label:
                cnt += 1


        if cnt >= len(label_in_video):
            target_demo_video.append(video)
```
